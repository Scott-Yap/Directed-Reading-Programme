Derivative: A financial contract that derives its value from the underlying.

Underlying: Financial assets
1. commodities: tradable physical goods
2. Bonds (loan)
3. Stocks ...

Option: contract that gives the holder the right, but no the obligation, to buy or sell an underlying asset at a predetermined price (strike price), on or before a specific date. 

Buy call: think underlying will rise
Buy Put: think underlying will drop

Eg: 
Buy call option to purchase Apple stocks at 150 pounds in 3 months
If Apple's price rise to 170, you can exercise the option, buy at 150, sell at 170 -> profit = 20 - premium
If price fall, let it expire, and just pay the premium

Key Terms:
Underlying: asset the option is based on 
Strike Price (K): agreed price to buy/sell the underlying
Expiry (T): the date the option expires
Premium: the price paid to buy the option
Call option: right to buy the asset at K
Put option: right to sell the asset at K


Payoff Diagrams (At Expiry)

Call Option (Buy Call)

$$
\text{Payoff} = \max(S_T - K, 0)
$$

- If $S_T > K$, you exercise (profit).  
- If $S_T \le K$, the option expires worthless.  

Put Option (Buy Put)

$$
\text{Payoff} = \max(K - S_T, 0)
$$

- If $S_T < K$, you exercise (profit).  
- If $S_T \ge K$, the option expires worthless.  

| Option Type | In the Money (ITM) | Out of the Money (OTM) |
|--------------|--------------------|-------------------------|
| Call | $S > K$ | $S < K$ |
| Put  | $S < K$ | $S > K$ |



Types of Options — with Examples

| Type | Description | Example |
|-------|--------------|----------|
| **European Option** | Exercised *only at expiry*. You decide whether to exercise at maturity based on the final price $S_T$. | You buy a **European call** on the S&P 500 with strike 4500 expiring in 1 month. <br>• If at expiry $S_T = 4700$, you gain $4700 − 4500 = 200$. <br>• If $S_T = 4300$, it expires worthless. |
| **American Option** | Can be exercised *anytime before expiry*. Useful when early exercise might be optimal (e.g. due to dividends). | You buy an **American call** on Apple at strike £150. <br>• If Apple announces a dividend soon and trades at £180, you might **exercise early** to receive the dividend. <br>• Otherwise, you typically wait until expiry. |
| **Asian Option** | Payoff depends on the *average* price of the underlying over time, not just the final price. | You buy an **Asian call** on oil with strike $80, based on the **average daily oil price** over the next month. <br>• If the average is $85$, payoff = $85 − 80 = $5. <br>• Helps smooth out short-term price spikes. |
| **Barrier Option** | Activated or deactivated if the underlying hits a certain barrier level. | You buy a **knock-out call** on gold with strike $2000$ and barrier $2100$. <br>• If gold ever hits $2100$, your option **expires immediately** (knocked out). <br>• If it never touches $2100$ and finishes above $2000$, you get the payoff. |
| **Digital / Binary Option** | Fixed payout if condition met; otherwise zero. Often used for *probability* or *event* trading. | You buy a **binary option** paying £100 if EUR/USD > 1.10 in 2 weeks, else £0. <br>• If it ends at 1.11 → you get £100. <br>• If 1.09 → nothing. <br>Price of the option ≈ implied probability × £100. |



Option greeks: measuring risk
- when price of stock moves
- when time passes
- when volatilty changes
- when interest rate shift

Option Greeks — Sensitivity Measures

The **Greeks** measure how an option’s price $V(S, t; r, \sigma, K, T)$ responds to changes in market parameters:

| Greek | Symbol | Measures sensitivity to... | Intuition |
|--------|---------|-----------------------------|------------|
| **Delta** | $\Delta = \frac{\partial V}{\partial S}$ | Changes in underlying price | Slope of the option price vs. stock price.<br>• Call: $0 \le \Delta \le 1$ <br>• Put: $-1 \le \Delta \le 0$ |
| **Gamma** | $\Gamma = \frac{\partial^2 V}{\partial S^2}$ | Changes in Delta (curvature) | Measures how fast Delta changes — the *convexity*. <br>High Gamma = sensitive to small price changes. |
| **Vega** | $\nu = \frac{\partial V}{\partial \sigma}$ | Changes in volatility | Tells how much the option value rises when volatility increases. <br>All options gain value when volatility ↑. |
| **Theta** | $\Theta = \frac{\partial V}{\partial t}$ | Passage of time | “Time decay.” Each day that passes, the option loses time value. <br>Usually: $\Theta < 0$ for long options. |
| **Rho** | $\rho = \frac{\partial V}{\partial r}$ | Changes in interest rate | Small effect. <br>Call ↑ when rates ↑, Put ↓ when rates ↑. |

---

Example Intuition (European Call)

| Greek | 
|--------|
| **Delta ≈ 0.6** → When stock rises £1, option price rises £0.60. |
| **Gamma = 0.05** → When stock rises £1, Delta increases by 0.05. |
| **Vega = 0.12** → 1% increase in volatility → option up £0.12. |
| **Theta = −0.03** → You lose £0.03 of time value per day. |
| **Rho = 0.08** → 1% rise in interest rate → option up £0.08. |
